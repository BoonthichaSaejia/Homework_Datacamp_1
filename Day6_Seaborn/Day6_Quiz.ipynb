```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Company=["A","B","C","D"]
Revenue=np.random.uniform(50,501,(4,4))
Year=[2014,2015,2016,2017]
df_1=pd.DataFrame(Revenue,Company,Year)
df_1
```

# 2. 

```python
sns.barplot(x=df_1.index,y=df_1.T.mean())
plt.xlabel("Company's Name")
plt.ylabel("Average Income")
```

```python
sns.barplot(y=df_1.index,x=df_1.T.mean())
plt.ylabel("Company's Name")
plt.xlabel("Average Income")
```

```python
sns.scatterplot(y=2017,x=df_1.index,data=df_1)
plt.xlabel("Company's Name")
plt.ylabel("Each Year Revenue (million dollar)")
```
