## Q13

```python
x=np.array(df['sqft_living']).reshape(-1,1)
y=np.array(df['sqft_living']).reshape(-1,1)
```

```python
from sklearn.preprocessing import StandardScaler
sc_x=StandardScaler()
sc_y=StandardScaler()

x_scaled=sc_x.fit_transform(x)
y_scaled=sc_y.fit_transform(y)
```

## Q12

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y_scaled,test_size=0.3,random_state=99)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

## Q14

### 1 Linear

```python
from sklearn.svm import SVR
```

```python
svr_linear=SVR(kernel='linear')
svr_linear.fit(x_train,y_train)
```

```python
linear_predicted=sc_y.inverse_transform(svr_linear.predict(x_test))
```

```python
linear_predicted
```

### 2 Radial Basis Function

```python
svr_rbf=SVR(kernel='rbf')
svr_rbf.fit(x_train,y_train)
```

```python
rbf_predicted=sc_y.inverse_transform(svr_rbf.predict(x_test))
```

```python
rbf_predicted
```

## Q15

```python
from sklearn import metrics
```

### 1 Linear

```python
print('MAE: ', metrics.mean_absolute_error(sc_y.inverse_transform(y_test), linear_predicted))
print('MSE: ', metrics.mean_squared_error(sc_y.inverse_transform(y_test), linear_predicted))
print('RMSE: ', np.sqrt(metrics.mean_squared_error(sc_y.inverse_transform(y_test), linear_predicted)))
```

### 2 RBF

```python
print('MAE: ', metrics.mean_absolute_error(sc_y.inverse_transform(y_test), rbf_predicted))
print('MSE: ', metrics.mean_squared_error(sc_y.inverse_transform(y_test), rbf_predicted))
print('RMSE: ', np.sqrt(metrics.mean_squared_error(sc_y.inverse_transform(y_test), rbf_predicted)))
```
