```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer=load_breast_cancer()
cancer
```

```python
print(cancer.DESCR)
```

```python
df=pd.DataFrame(cancer.data,columns=cancer.feature_names)
df
```

```python
label=cancer.target
```

```python
len(label)
```

```python
df['label']=label
```

```python
df.sample(10)
```

```python
df['label_name']=['malignant' if x==0 else 'benign' for x in df['label']]
```

```python
df.info()
```

```python
df.describe()
```

```python
df.corr()
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='mean radius',y='worst symmetry',data=df)
```

```python
df
```

```python
df2=df.drop(['label_name'],axis=1)
```

```python
df2
```

## 1 Default

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(df2.drop(['label'],axis=1),
                                              df2['label'],
                                              test_size=0.5,
                                              random_state=1)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
nb=GaussianNB()
nb.fit(X_train,y_train)
```

```python
predicted=nb.predict(X_test)
predicted
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, precision_score, recall_score
```

```python
confusion_matrix(y_test,predicted)
```

```python
print('Accuracy=',accuracy_score(y_test,predicted))
print('F1 score=',f1_score(y_test,predicted))
print('Precision=',precision_score(y_test,predicted))
print('Recall=',recall_score(y_test,predicted))
```
