## Q15

#### 1 Normalization

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
min_max_scaler=MinMaxScaler()
```

```python
df_minmax=min_max_scaler.fit_transform(df.drop(['Type'],axis=1))
df_minmax
```

```python
norm_df=pd.DataFrame(df_minmax,columns=df.columns[:-1])
```

```python
norm_df
```

```python
X_train,X_test,y_train,y_test=train_test_split(norm_df,
                                              df['Type'],
                                              test_size=0.2,
                                              random_state=99)
```

```python
from sklearn.naive_bayes import GaussianNB
nb14_1=GaussianNB()
nb14_1.fit(X_train,y_train)
```

```python
predicted14_1=nb14_1.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, precision_score, recall_score
confusion_matrix(y_test,predicted14_1)
```

```python
print('Accuracy=',accuracy_score(y_test,predicted14_1))
print('F1 score=',f1_score(y_test,predicted14_1,average='micro'))
print('Precision=',precision_score(y_test,predicted14_1,average='micro'))
print('Recall=',recall_score(y_test,predicted14_1,average='micro'))
```

```python
before_norm_f1=f1_score(y_test,predicted14_1,average='micro')
before_norm_f1
```
