## Q15

```python
from sklearn.preprocessing import MinMaxScaler
min_max_scaler=MinMaxScaler()
```

```python
df_minmax=min_max_scaler.fit_transform(df.drop('Wine class labels',axis=1))
```

```python
new_df=pd.DataFrame(df_minmax,columns=df.columns[1:]) # don't want the last two columns
new_df
```

## Q16

```python
from sklearn.neighbors import KNeighborsClassifier
X_train, X_test, y_train, y_test=train_test_split(new_df,df['Wine class labels'], test_size=0.30,random_state=69)
```

```python
knn= KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train,y_train)
```

```python
predicted=knn.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, precision_score, recall_score
confusion_matrix(y_test,predicted)
```

```python
print('Accuracy=',accuracy_score(y_test,predicted))
print('F1 Score=',f1_score(y_test,predicted,average='micro'))
print('Accuracy=',precision_score(y_test,predicted,average='micro'))
print('Recall=',recall_score(y_test,predicted,average='micro'))
```

## Q17

```python
accuracy_lst=[]
for i in range(1,50):
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    predicted_i=knn.predict(X_test)
    accuracy_lst.append(accuracy_score(y_test,predicted_i))
```

```python
accuracy_lst
```

## Q19

```python
plt.figure(figsize=(10,10))
plt.plot(range(1,50),accuracy_lst,color='black',linestyle='dashed',marker='o', 
         markerfacecolor='blue', markersize=7)
plt.xlabel('Number of K')
plt.ylabel('Accuracy')
```
