```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test=train_test_split(df.drop('Wine class labels',axis=1),df['Wine class labels'], test_size=0.4,random_state=69)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

## Q16

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
x_train
```

```python
x_test
```

## Q17

```python
from sklearn.svm import SVC
```

```python
svc=SVC()
```

```python
svc.fit(x_train,y_train)
```

```python
svc_predicted=svc.predict(x_test)
```

## Q18

```python
svc_predicted
```

```python
sns.countplot(x=svc_predicted)
```

## Q19

```python
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, precision_score, recall_score
```

```python
confusion_matrix(y_test,svc_predicted)
```

## Q20

```python
svc_acc_ma   =accuracy_score(y_test,svc_predicted)
svc_f1_ma    =f1_score(y_test,svc_predicted,average='macro')
svc_pre_ma   =precision_score(y_test,svc_predicted,average='macro')
svc_recall_ma=recall_score(y_test,svc_predicted,average='macro')
```

```python
svc_acc_mi   =accuracy_score(y_test,svc_predicted)
svc_f1_mi    =f1_score(y_test,svc_predicted,average='micro')
svc_pre_mi   =precision_score(y_test,svc_predicted,average='micro')
svc_recall_mi=recall_score(y_test,svc_predicted,average='micro')
```

```python
svc_df=pd.DataFrame(np.array([[svc_acc_ma,svc_acc_mi],[svc_f1_ma,svc_f1_mi],[svc_pre_ma,svc_pre_mi],[svc_recall_ma,svc_recall_mi]]),
                    columns=['Macro','Micro'],index=['Accuracy','F1','Precision','Recall'])
svc_df
```
