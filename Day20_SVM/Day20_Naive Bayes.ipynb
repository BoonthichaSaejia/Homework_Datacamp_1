## divide data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test,y_train,y_test=train_test_split(df.drop(['result','result_title'],axis=1),
                                               df['result'],test_size=0.3,random_state=20)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
X_train.head()
```

## create model

```python
from sklearn.svm import SVC
```

```python
svc=SVC()
```

```python
svc.fit(X_train,y_train)
```

```python
predicted=svc.predict(X_test)
```

```python
predicted
```

## evaluate model

```python
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, precision_score, recall_score
```

```python
confusion_matrix(y_test,predicted)
```

00 true negatives

10 false negatives

11 true positive

01 false positive

```python
print('Accuracy=',accuracy_score(y_test,predicted))
print('F1 score=',f1_score(y_test,predicted))
print('Precision=',precision_score(y_test,predicted))
print('Recall=',recall_score(y_test,predicted))
```

## tune parameter

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_combination = {'C': [0.01,0.1,1,10,100,100,1000,10000], #size of regulation
                    'gamma': [0.00001,0.0001,0.001,0.01,0.1,1,10]}
```

C มาก ลงโทษในกรณีข้ามฝั่งมาก so decision boundary จะแคบ

gamma มาก dicision boundary จะลดมาก มีความเป็นเส้นตรงน้อยลง sensitive กับ data มากขึ้น

kernal trick  มีผลต่อรูปร่างของ hyperplane and decision boundary
default is RBF (โค้งตามข้อมูล) which is not linear


```python
grid_search=GridSearchCV(SVC(),param_combination,verbose=1)
```

```python
grid_search.fit(X_train,y_train)
```

```python
grid_search.best_params_
```

```python
grid_search.best_estimator_
```

```python
grid_predicted=grid_search.predict(X_test)
```

```python
confusion_matrix(y_test,grid_predicted)
```

```python
print('Accuracy=',accuracy_score(y_test,grid_predicted))
print('F1 score=',f1_score(y_test,grid_predicted))
print('Precision=',precision_score(y_test,grid_predicted))
print('Recall=',recall_score(y_test,grid_predicted))
```

ค่า accuracy สูงขึ้น
