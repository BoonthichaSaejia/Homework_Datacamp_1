```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x1=np.arange(0,31)
y1=x1**2
plt.plot(x1,y1)
plt.title("The x power 2 function")
plt.xlabel("Number 1-30")
plt.ylabel("X axis power 2")
```

# 2.

```python
x2=np.arange(0,31)
y2=np.sin(x2)
plt.plot(x2,y2,c="pink")
plt.title("The sine graph")
plt.xlabel("Number 1-30")
plt.ylabel("sinx")
plt.show()
```

# 3.

```python
y3=np.arange(0,31)
x3=y3**2
plt.plot(x3,y3,c="#00FF00")
plt.title("The power 2 of y graph")
plt.xlabel("Number 1-30")
plt.ylabel("power y")
plt.show()
```

# 4.

```python
x4=np.arange(0,31)
y4=np.cos(x4)
plt.plot(x4,y4,c="red",ls="--")
plt.title("The cosine graph")
plt.xlabel("Number 1-30")
plt.ylabel("cosx")
plt.show()
```

# 5.

```python
x5=np.arange(0,11)
y5_1=x5**2
y5_2=x5**3
plt.plot(x5,y5_1,c="red",marker="o")
plt.plot(x5,y5_2,c="blue",marker="^")
plt.title("The power 2 and power 3 graph")
plt.xlabel("Number 1-10")
plt.ylabel("value")
plt.show()
```

# 6.

```python
x6_1=np.arange(0,6)
y6_1=x6_1**2
y6_2=np.arange(0,6)
x6_2=y6_2**2
plt.plot(x6_1,y6_1,c="red",marker="o",label="y=x^2")
plt.plot(x6_2,y6_2,c="blue",marker="^",label="x=y^2")
plt.legend(loc=3)  
plt.grid()
plt.title("The graph")
plt.xlabel("value x")
plt.ylabel("value y")
plt.show()
```

# 7.

```python
y7=np.arange(0,20)
x7=(y7**2)+(y7*4)
plt.plot(x7,y7,c="#fffd00",marker="o")
plt.grid()
plt.title("The exponential graph")
plt.xlabel("y7**2+y7*4")
plt.ylabel("y")
plt.xlim([0,400])
plt.ylim([0,17.5])
plt.show()
```

# 8.

```python
y8_1=np.arange(0,10)
x8_1=y8_1**3
x8_2=np.arange(0,10)
y8_2=x8_2**2+2
fig=plt.figure(figsize=(6,6))
ax1=fig.add_axes([0.1,0.1,1,1])
ax1.plot(x8_1,y8_1,"r",label="y**3")
plt.title("power 3 of y Graph")
plt.xlabel("value x")
plt.ylabel("value y")
plt.legend(loc=0)  
ax2=fig.add_axes([0.6,0.2,0.4,0.4])
ax2.plot(x8_2,y8_2,"b",label="x**2+2")
plt.title("power 2 of x Graph")
plt.xlabel("value x")
plt.ylabel("value y")
plt.legend(loc=0)  
plt.show()
```
