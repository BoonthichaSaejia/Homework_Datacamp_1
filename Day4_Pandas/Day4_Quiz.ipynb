# Quiz20

```python
df_titanic['Age'].max()
```

```python
df_titanic_20_1=df_titanic[(df_titanic['Age']>0)&(df_titanic['Age']<=10)]
df_titanic_20_2=df_titanic[(df_titanic['Age']>10)&(df_titanic['Age']<=20)]
df_titanic_20_3=df_titanic[(df_titanic['Age']>20)&(df_titanic['Age']<=30)]
df_titanic_20_4=df_titanic[(df_titanic['Age']>30)&(df_titanic['Age']<=40)]
df_titanic_20_5=df_titanic[(df_titanic['Age']>40)&(df_titanic['Age']<=50)]
df_titanic_20_6=df_titanic[(df_titanic['Age']>50)&(df_titanic['Age']<=60)]
df_titanic_20_7=df_titanic[(df_titanic['Age']>60)&(df_titanic['Age']<=70)]
df_titanic_20_8=df_titanic[(df_titanic['Age']>70)&(df_titanic['Age']<=80)]
```

```python
Survived_1=(df_titanic_20_1['Survived']==1).sum()
Survived_2=(df_titanic_20_2['Survived']==1).sum()
Survived_3=(df_titanic_20_3['Survived']==1).sum()
Survived_4=(df_titanic_20_4['Survived']==1).sum()
Survived_5=(df_titanic_20_5['Survived']==1).sum()
Survived_6=(df_titanic_20_6['Survived']==1).sum()
Survived_7=(df_titanic_20_7['Survived']==1).sum()
Survived_8=(df_titanic_20_8['Survived']==1).sum()
```

```python
len(df_titanic_20_1)
len(df_titanic_20_2)
len(df_titanic_20_3)
len(df_titanic_20_4)
len(df_titanic_20_5)
len(df_titanic_20_6)
len(df_titanic_20_7)
len(df_titanic_20_8)
```

```python
print('Survival rate of 0-10 year-old people is',Survived_1/len(df_titanic_20_1))
print('Survival rate of 10-20 year-old people is',Survived_2/len(df_titanic_20_2))
print('Survival rate of 20-30 year-old people is',Survived_3/len(df_titanic_20_3))
print('Survival rate of 30-40 year-old people is',Survived_4/len(df_titanic_20_4))
print('Survival rate of 40-50 year-old people is',Survived_5/len(df_titanic_20_5))
print('Survival rate of 50-60 year-old people is',Survived_6/len(df_titanic_20_6))
print('Survival rate of 60-70 year-old people is',Survived_7/len(df_titanic_20_7))
print('Survival rate of 70-80 year-old people is',Survived_8/len(df_titanic_20_8))
```
