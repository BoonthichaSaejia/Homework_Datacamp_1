## Q15-Q19

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
import string
```

```python
def text_process(mess):
    no_punc=[x for x in mess if x not in string.punctuation]
    
    no_punc=''.join(no_punc)
    
    text=[word for word in no_punc.split() if word.lower() not in stopwords.words('english')]
    
    return text
```

```python
bow_transformer=CountVectorizer(analyzer=text_process).fit(df['text'])
bow_transformer
```

```python
bow_sms=bow_transformer.transform(df['text'])
print('Shape of Sparse Matrix', bow_sms.shape)
print('None Zero', bow_sms.nnz)
print('% of Sparsity:',bow_sms.nnz*100/(bow_sms.shape[0]*bow_sms.shape[1]))
```

## Q20

```python
from sklearn.feature_extraction.text import TfidfTransformer
```

```python
tfidf_transformer=TfidfTransformer().fit(bow_sms)
tfidf_transformer
```

```python
tfidf_sms=tfidf_transformer.transform(bow_sms)
tfidf_sms
```

## Q21

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
nb=MultinomialNB()
```

```python
y=df['stars']
```

```python
x=tfidf_sms
```

```python
star_detection=nb.fit(x,y)
star_detection
```

```python
predicted = star_detection.predict(x)
predicted
```

## Q22

```python
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, precision_score, recall_score
```

```python
confusion_matrix(y,predicted)
```

```python
acc_ma   =accuracy_score(y,predicted)
f1_ma    =f1_score(y,predicted,average='macro')
pre_ma   =precision_score(y,predicted,average='macro')
recall_ma=recall_score(y,predicted,average='macro')
```

```python
print(acc_ma)
print(f1_ma)
print(pre_ma)
print(recall_ma)
```

## Q23

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test= train_test_split(df['text'],df['stars'],test_size=0.2)
```

## Q24

```python
from sklearn.pipeline import Pipeline
```

```python
pipeline=Pipeline([
    ('bow',CountVectorizer(analyzer=text_process)),
    ('tfidf',TfidfTransformer()),
    ('classifier', MultinomialNB())
])
```

```python
pipeline.fit(x_train,y_train)
```

```python
predicted = pipeline.predict(x_test)
```

## Q25

```python
confusion_matrix(y_test,predicted)
```

```python
split_acc_ma   =accuracy_score(y_test,predicted)
split_f1_ma    =f1_score(y_test,predicted,average='macro')
split_pre_ma   =precision_score(y_test,predicted,average='macro')
split_recall_ma=recall_score(y_test,predicted,average='macro')
```

```python
print(split_acc_ma)
print(split_f1_ma)
print(split_pre_ma)
print(split_recall_ma)
```
