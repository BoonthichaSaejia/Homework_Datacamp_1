## Q9

```python
import scipy.cluster.hierarchy as sch
```

```python
fig=plt.figure(figsize=(14,8))
dendrogram = sch.dendrogram(sch.linkage(df[['STG','PEG']],method='ward'))
plt.title('Dandrogram')
plt.xlabel('Student')
plt.ylabel('Euclidean distances')
```

The best is 2 cluster but 4 clusters is also acceptable

## Q10 & Q11

```python
from sklearn.cluster import AgglomerativeClustering
hc=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
```

```python
hc_10=hc.fit_predict(df[['STG','PEG']])
```

```python
hc_10
```

```python
df[['STG','PEG']][hc_10==0]
```

```python
x10_1=np.array(df[['STG','PEG']][hc_10==0])[:,0]
y10_1=np.array(df[['STG','PEG']][hc_10==0])[:,1]
x10_2=np.array(df[['STG','PEG']][hc_10==1])[:,0]
y10_2=np.array(df[['STG','PEG']][hc_10==1])[:,1]
x10_3=np.array(df[['STG','PEG']][hc_10==2])[:,0]
y10_3=np.array(df[['STG','PEG']][hc_10==2])[:,1]
```

```python
fig=plt.figure(figsize=(12,8))
plt.scatter(x10_1,y10_1,s=35,c='green',label='Cluster1')
plt.scatter(x10_2,y10_2,s=35,c='red',label='Cluster2')
plt.scatter(x10_3,y10_3,s=35,c='blue',label='Cluster3')
plt.xlabel('STG')
plt.ylabel('PEG')
plt.legend()
plt.grid
```
