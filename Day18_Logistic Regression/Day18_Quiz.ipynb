```python
def plotMultiDistplot(df):
    count=0
    colList=[]
    for x in df.columns:
        if (df[x].dtype=='float64' or df[x].dtype =='int64') and len(df[x].unique())>2 : 
            count+=1
            colList.append(x)
    row=round(count/2)
    fig, axs = plt.subplots(row, 2,figsize=(18,row*4),gridspec_kw={'hspace':0.4, 'wspace': 0.2})
    for i in range(row):
        for j in range(2):
            if len(colList)!=0:
                colName=colList.pop()
                sns.distplot(ax=axs[i,j],a=df[colName])
                axs[i,j].set_title(colName)
```

```python
plotMultiDistplot(df)
```

```python
def plotMultiHistogram(df):
    count=0
    colList=[]
    for x in df.columns:
        if (df[x].dtype=='float64' or df[x].dtype =='int64') and len(df[x].unique())>2 : 
            count+=1
            colList.append(x)
    row=round(count/2)
    fig, axs = plt.subplots(row, 2,figsize=(18,row*4),gridspec_kw={'hspace':0.4, 'wspace': 0.2})
    for i in range(row):
        for j in range(2):
            if len(colList)!=0:
                colName=colList.pop()
                axs[i,j].hist(df[colName])
                axs[i,j].set_title(colName)
```

```python
plotMultiHistogram(df)
```

## Q11

```python
import math
def plotMultiBoxplot(df):
    count=0
    colList=[]
    for x in df.columns:
        if (df[x].dtype=='float64' or df[x].dtype =='int64') and len(df[x].unique())>2 : 
            count+=1
            colList.append(x)
    col=math.ceil(count/4)
    fig, axs = plt.subplots(4, col,figsize=(18,col*6),gridspec_kw={'hspace':0.4, 'wspace': 0.2})
    for i in range(4):
        for j in range(col):
            if len(colList)!=0:
                colName=colList.pop()
                sns.boxplot(ax=axs[i,j],y=df[colName],x=df['classification'])
                axs[i,j].set_title(colName)
```

```python
plotMultiBoxplot(df)
```
